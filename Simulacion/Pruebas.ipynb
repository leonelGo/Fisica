```python
# Datos para graficar
x = np.linspace(0, 2*np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Primera figura con subplots 2x2
fig1, (ax1, ax2) = plt.subplots(2, figsize=(8, 6))
ax1[0].plot(x, y1, color='blue')
ax1[0].set_title('Seno (Figura 1)')
ax1[1].plot(x, y2, color='red')
ax1[1].set_title('Coseno (Figura 1)')
fig1.tight_layout()

# Segunda figura con subplots 2x2
fig2, ax2 = plt.subplots(2, 2, figsize=(8, 6))
ax2[0, 1].plot(x, y1, color='green')
ax2[0, 1].set_title('Seno (Figura 2)')
ax2[1, 1].plot(x, y2, color='orange')
ax2[1, 1].set_title('Coseno (Figura 2)')
fig2.tight_layout()

plt.show()
```
